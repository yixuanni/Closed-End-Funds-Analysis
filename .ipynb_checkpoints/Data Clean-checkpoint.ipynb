```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import datetime
```

```python
cef_daily = pd.read_csv('cef_daily.csv')
cef_daily
```

```python
cef_list = pd.read_csv('cef_list.csv')
cef_list
```

```python
cef_list = cef_list[['Ticker', 'Class', 'ExpenseRatio']]
cef_list
```

```python
cef_daily_n = cef_daily.drop(columns=['return.mkt', 'symbol.nav', 'return.nav'], axis = 1)
cef_daily_n
```

```python
cef_daily_n['date'] = pd.to_datetime(cef_daily_n['date'])
cef_daily_n = cef_daily_n[cef_daily_n['date'] >= '2019-01-01']
cef_daily_n
```

```python
cef_daily_n.isna().sum()
```

```python
rows_with_na = cef_daily_n[cef_daily_n.isna().any(axis=1)]
rows_with_na
```

```python
rows_with_na.symbol.unique()
```

```python
pd.set_option('display.max_rows', None)
rows_with_na
```

```python
pd.reset_option('display.max_rows')
```

```python
cef_daily_n = cef_daily_n[~cef_daily_n['symbol'].isin(['NID', 'NIQ'])]
cef_list = cef_list[~cef_list['Ticker'].isin(['NID', 'NIQ'])]
cef_daily_n.isna().sum()
```

```python
cef_list = cef_list.set_index(['Ticker'])
cef_daily1 = cef_daily_n.set_index(['symbol'])
cef_daily1.date = pd.to_datetime(cef_daily1.date)
cef_daily1.dtypes
```

```python
cef_daily1
```

```python
cef = pd.merge(cef_daily1, cef_list, left_index=True, right_index=True)
cef
```

```python
cef.set_index('date', append=True, inplace=True)
cef
```

```python
cef.isna().sum()
```

```python
cef.to_csv("cef_clean.csv")
```
